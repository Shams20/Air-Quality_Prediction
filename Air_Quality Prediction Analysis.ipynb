```python
# Create the dummies variable

df['AQI_Bucket']=pd.get_dummies(df['AQI_Bucket'], drop_first=True)
```

```python
df.head(5)
```

# # Preparing Data

```python
# In this step we devide the data into into attributes(input), labels(outputs)

x = df.drop(['City','Date','AQI_Bucket'], axis=1)
y = df.AQI_Bucket
```

# Spliting Data into train & test

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
x.head(2)
```

```python
# Here we see the shape of x and y

print('shape of x_train',x_train.shape)
print('Shape of y_train: ',y_train.shape)

print('\n')

print('Shape of X_test: ',x_test.shape)
print('Shape of y_test: ',y_test.shape)
```

# Training Algorithm

we have split our data into train & test now time to train our model

# Linear Regression Algorithm

```python
from sklearn.linear_model import LinearRegression

regressor=LinearRegression()
regressor.fit(x_train,y_train)  # Fitting the model

print("Training Complete")
```

```python
# To retrive intercept & Coefficient

print("intercept is :")
print(regressor.intercept_)
```

```python
print("coefficient is :")
print(regressor.coef_)
```

```python
print(x_test)   

y_pred= regressor.predict(x_test)
```

```python
# comparing Actual v/s Predicted

df=pd.DataFrame({'Actual':y_test,'Predicted':y_pred})
df
```

```python
plt.figure(figsize=(14,6))

df.plot(kind='bar', figsize=(12,6), cmap="Set1")

plt.title('Actual vs Predicted score', fontsize=22)  
plt.tight_layout()
```

```python
from sklearn import metrics

print('Mean absolute error:',metrics.mean_absolute_error(y_test,y_pred))
print('Mean Squared error:',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Square error:', np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

# Logistic Regression model

```python
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression

logR=LogisticRegression()

logR.fit(x_train,y_train)

predictR= logR.predict(x_test)
print("")

print('Accuracy_score of Logistic regression:', accuracy_score(y_test,predictR)*100)

print('classification report of Logistic Regression:')
print("")

print(classification_report(y_test,predictR))


```
